```python
def show_data(data):
    if isinstance(data, str):
        print(f"String: {data}")    
    elif isinstance(data, int):
        print(f"Integer: {data}")
    elif isinstance(data, list):
        print(f"List: {data}")
    elif isinstance(data, dict):
        print(f"Dictionary: {data}")    
    else:        
        print("Unknown data type")
```

```python
show_data("Hello, World!")
```

```python
show_data([1, 2, 3, 4, 5])
```

```python
show_data({"name": "Alice", "age": 30})
```

```python
show_data(77)
```

```python
from functools import singledispatch
```

```python
@singledispatch
def show_data(data):
    print(f"Unknown data type: {type(data)}")

@show_data.register
def _(data: str):
    print(f"String: {data}")

@show_data.register
def _(data: int):
    print(f"Integer: {data}")

@show_data.register
def _(data: list):
    print(f"List: {data}")

@show_data.register
def _(data: dict):
    print(f"Dictionary: {data}")
```
