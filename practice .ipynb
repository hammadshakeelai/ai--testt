### Exercise 1.1: Data Structures and Filtering
**Task:** 1. Create a list of dictionaries where each dictionary represents a student with `name` and `score`.
2. Use a `for` loop or the `filter()` function to extract only the students who scored 60 or above.
3. Print the names of the passing students.

```python
# 1. Define your list of dictionaries here
students = [
    {"name": "Ali", "score": 85},
    {"name": "Omar", "score": 45},
    {"name": "Ayesha", "score": 78},
    {"name": "Bilal", "score": 55}
]

# 2. Write a function to check if a student passed (score >= 60)
def check_pass(student):
    # YOUR CODE HERE: Return True if score >= 60, else False
    return (student['score'] >= 60)

# 3. Use filter() or a loop to get passed students
# YOUR CODE HERE
passed_student = []
for student in students:
    if check_pass(student):
        passed_student.append(student['name'])
# 4. Print the result
# YOUR CODE HERE
passed_student
```

### Exercise 2.1: Flexible Functions
**Task:**
1. Write a function called `calculate_total` that takes an arbitrary number of numeric arguments using `*args` and returns their sum.
2. Create a lambda function that squares a given number. Use the `map()` function to apply this lambda to a list of numbers.

```python
# 1. Define calculate_total using *args
def calculate_total(*args):
    # YOUR CODE HERE: Sum the arguments and return the result
    sum = 0
    for i in args:
        sum+=i
    return sum

# Test your function
print("Total Sum:", calculate_total(10, 20, 30, 40.5))

# 2. Lambda and Map
numbers = [2, 4, 6, 8]

# YOUR CODE HERE: Create a lambda function to square a number
# square_func = lambda ...
square_func = lambda x : x**2 

# YOUR CODE HERE: Use map() to apply the lambda to the 'numbers' list
# squared_numbers = map(...)
print(map( square_func ,numbers))
# Print the list of squared numbers
# print(list(squared_numbers))
```
